```python
def selection(self, population):
    '''
    Селекция методом рулетки
    '''
    full_estimate = sum([ind.estimate for ind in population])
    for individual in population:
        individual.percent = 1 / (1 + individual.estimate) * 100
        
    wheel = list(it.accumulate([individual.percent for individual in population]))[:-1]
    wheel.insert(0, 0)
    
    population = random.choices(
        population, 
        weights=[individual.percent for individual in population],
        k=len(population)
    )
    
    return population
```
